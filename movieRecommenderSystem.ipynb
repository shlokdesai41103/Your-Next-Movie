```python
import numpy as  np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
 movies = movies.merge(credits, on='title')
```

```python
movies.info()
```

```python
# What factors do we want to take into account when recommending movies?
# genres
# id
# keywords
# title
# overview
# cast
# crew
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
#change the current format of genres from:
#[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
#to:
#['Action', 'Adventure', 'Fantasy', 'SciFi']
```

```python
import ast
#ast is going to help us convert a list which is in the form of a string to a list that we can use as a regular list
def convertGenres(obj):
    genres = []
    for i in ast.literal_eval(obj):
        genres.append(i["name"])
    return genres
```

```python
#make the genres column simpler and more usable
movies['genres'] = movies['genres'].apply(convertGenres)
```

```python
movies.head()
```

```python
#make the genres column simpler and more usable
movies['keywords'] = movies['keywords'].apply(convertGenres)
```

```python
movies.head()
```

```python
#Now to do the same with cast
#We only want to take the most important actors, we want to simplify the cast column by only including the name of the top 3 actors
def convertCast(obj):
    cast = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            cast.append(i["name"])
            count+=1
        else:
            break
    return cast
```

```python
movies['cast'] = movies['cast'].apply(convertCast)
```

```python
movies.head()
```

```python
#Now to process the crew column
#We only care about the dictionary that contains the name of the director
#We need to make a function that filters all except the director
def getDirector(obj):
    director = []
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            director.append(i["name"])
    return director
```

```python
movies['crew'] = movies['crew'].apply(getDirector)
```

```python
movies.head()
```

```python
#concatenate the overview of the movies
#We want to make the overview usable too by converting it into a list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#Our recommendation system is based on tags
#tags are words without spaces in between
#So we have to remove spaces between words that should make up one characteristic or a tag
#Eg: Sam Worthington would make two tags Sam and Worthington. So someone that watched avatar would have wathched a movie with a tag or Sam and Worthington.
#Another movie Spectre has the director Sam Mendes which would make tag Sam and Mendes.
# Due to the similar tags it is very likely that someone that has watched Avatar will be recommended Spectre due to the common Sam tag.
#To fight this, we have to concatenate strings to form tags: Sam Worthington -> SamWorthington, Sam Mendes -> SamMendes
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies.head()
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['cast']+movies['keywords']+movies['crew']
movies.head()
```

```python
#As we've compiled all columns into one tag column, we now create a new data frame with only tthe necessary columns
new_df = movies[['movie_id','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
#All the tags have been compiled into large strings which are easily searchable
new_df['tags'][0]
new_df.head()
```

```python
#We will now work towards vectorizing our data
#We will be doing text vectorization using bag of words technique
#Dont consider stop words during vectorization (are, and, if, from, to....)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
#There are some flaws with our features. Similar words have created different tags. Eg: active and activity are two different tags with a similar meaning.
#We can prevent this using stemming, the nltk library
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
#When vectorizing the data we wont be calculating the euclidian distance.
#We will be calculating the cosine distance. Meaning the angle between two vectors. 
#If angle is very small, then they are basically the same vector
#The higher dimension you work with, the more inaccurate euclidian distance gets, cosine is better there.
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
sorted(similarity[0], reverse=True)
```

```python
similarity[0]#Similarity of the first movie with every other movies in the database
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
